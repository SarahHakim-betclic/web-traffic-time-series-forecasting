```python
import tensorflow.keras.backend as K

def smape_loss(y_true, y_pred):
    epsilon = 0.001
    summ = K.maximum(K.abs(y_true) + K.abs(y_pred) + epsilon, 0.5 + epsilon)
    smape = K.abs(y_pred - y_true) / summ * 2.0
    return smape

# SMAPE as Kaggle calculates it
def kaggle_smape(true, predicted):
    true_o = true
    pred_o = predicted
    summ = K.abs(true_o) + K.abs(pred_o) + 0.000001
    smape = np.where(summ==0, 0, K.abs(pred_o - true_o) / summ)
    return smape
```

```python
# Initialising the RNN
regressor = Sequential()

# Adding the input layerand the LSTM layer
regressor.add(LSTM(units = 8, activation = 'relu', input_shape = (None, 1)))

# Adding the output layer
regressor.add(Dense(units = 1))

# Compiling the RNN
regressor.compile(optimizer = 'adam', loss = smape_loss)

# Fitting the RNN to the Training set
regressor.fit(x_train, y_train, batch_size = 512, epochs = 10, verbose = 1, validation_split=0.1)
```
